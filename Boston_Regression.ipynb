## ML with Load_Boston()
1. **import python packages**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn import linear_model
# from sklearn.linear_model import LinearRegression
```

2. **Load data Boston**

```python
dataBoston = load_boston()
# print(dir(dataBoston))  #['DESCR', 'data', 'feature_names', 'filename', 'target']
# print(dataBoston['data'])
# print(dataBoston['data'].shape) #dimensi
# print(dataBoston['data'][0])
# print(dataBoston['feature_names'])
# print(dataBoston['target'])

df=pd.DataFrame(
    dataBoston['data'],
    columns=dataBoston['feature_names']
)

df['price']=dataBoston['target']
df
```

3. **Korelasi Data**

```python
korelasi = df.corr()
plt.imshow(korelasi, cmap='BuPu_r')
plt.colorbar()
# df.columns
plt.xticks(np.arange(14),list(df.columns))
plt.yticks(np.arange(14),list(df.columns))
plt.show()
```

4. **Regresi Linear**

```python
model=linear_model.LinearRegression()
model.fit(df[['ZN','CHAS','RM','DIS', 'B']], df['price'])
model.coef_,model.intercept_
model.predict(df[['ZN','CHAS','RM','DIS', 'B']])
# df.head(1)
# df.head(1)[['ZN','CHAS','RM','DIS', 'B']]
model.predict(df.head(1)[['ZN','CHAS','RM','DIS', 'B']])
```
